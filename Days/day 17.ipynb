```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Types of graph
1. plot             line graph, between two parameters
2. bar              bar graph of single parameter(optional cat_data)
3. hist             histogram of a bunch of data i.e. frequency of single pattern 
4. scatter          dot graph between two parameters 
5. boxplot/box      statistical overview in form of box for a single parameter- mean,outliers,quantiles(25%, 75%)IQR (optional- categorical data)
6. stem             more like bar graph, lines instead of bar 
7. violinplot       violin shaped statistical overview graph & spread of data ( more like histogram) &single parameter (optional cat-data)

```python
a=np.linspace(0,5,11)
x=np.arange(1,8)
x2=np.arange(1,8)
```

```python
x
```

```python
fig=plt.figure(figsize=(5,10))
y=x**3
y2=x2 **4
plt.plot(x,y,x2,y2,lw=3,marker='o',markersize=4)
plt.legend(["y=x**3",'y2=x2**4'])
plt.grid()
plt.show()
fig.savefig('graph.png')
```

```python
plt.bar(x,y,width=0.5)
plt.title('Bar Graph')
plt.legend("hii")
plt.show()
```

```python
plt.hist(x,y)
```

```python
plt.scatter(x,y)
```

```python
plt.boxplot(x)
plt.show()
```

```python
plt.stem(x,y)
plt.xlabel('x==')
plt.ylabel('y==')
plt.title('stem')
plt.grid()
plt.xticks(x)
plt.yticks(range(0,200,10))
plt.show()
```

```python
plt.violinplot(x)
plt.title('violinplot')
```
